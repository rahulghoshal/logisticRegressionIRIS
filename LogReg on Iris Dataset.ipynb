```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Dataset/Iris.csv')
```

```python
data.head()
```

Extracting feature columns from the original "Iris" dataset

```python
features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
```

Creating the target column [Iris-setosa : 0, Iris-versicolor : 1, Iris-virginica : 2]

```python
data['target'] = data['Species'].replace({'Iris-setosa' : 0, 'Iris-versicolor' : 1, 'Iris-virginica' : 2})
```

Printing the features and target

```python
X = data[features]
y = data['target']
```

```python
print(X)
print(y)
```

Splitting the dataset into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=50)
```

Training the Logistic Regression model using training datasets

```python
logreg_model = LogisticRegression()
logreg_model.fit(X_train,y_train)
```

Predicting on Testing Data 

```python
logreg_pred = logreg_model.predict(X_test)
```

```python
logreg_pred
```

Model Accuracy

```python
logreg_accuracy = accuracy_score(y_test, logreg_pred)
print(logreg_accuracy)
```

Checking on a sample data and making predictions using Species Name 

```python
sample_data = [[6.5,3.0,5.2,2.0]]
sample_prediction = logreg_model.predict(sample_data)

if (sample_prediction==0):
    print("Flower is from Iris-setosa species")
elif (sample_prediction==1):
    print("Flower is from Iris-versicolor species")
elif (sample_prediction==2):
    print("Flower is from Iris-virginica species")
```
